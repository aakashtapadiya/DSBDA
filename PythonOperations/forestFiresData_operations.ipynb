```python
sub1 = df[(df['area']>=0 ) & (df['area']<363)]
```

```python
sub1
```

```python
sub2 = df[(df['area']>=363 ) & (df['area']<1110)]
```

```python
sub2
```

### Merge subset

```python
sub3 = pd.merge(sub2, sub1, how='outer')
```

```python
sub3
```
