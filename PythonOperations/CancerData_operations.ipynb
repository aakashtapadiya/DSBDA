```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('breast-cancer.csv')
df
```

### Data cleaning

```python
df.isna().sum()
```

### Outlier removal

```python
print('Highest value' , df.radius_mean.mean()+3*df.radius_mean.std())
```

```python
print('Lowest value' , df.radius_mean.mean()-3*df.radius_mean.std())
```

```python
df[(df.radius_mean>24)|(df.radius_mean<3)]
```

```python
df = df[(df.radius_mean<24)&(df.radius_mean>3)]
```

```python
df
```

### Data transformation

```python
df.area_worst = df.area_worst.apply(lambda x:round(x) )
df.area_worst
```

### Model Building(Naive Bayes)

```python
x = df.drop(['id', 'diagnosis'], axis=1)
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
lb.fit(df.diagnosis)
```

```python
y = lb.transform(df.diagnosis)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr, x_te, y_tr, y_te = train_test_split(x,y, train_size=0.75)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_tr,y_tr)
```

```python
y_pred = model.predict(x_te)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_pred, y_te)
```

```python
confusion_matrix(y_pred,y_te)
```
