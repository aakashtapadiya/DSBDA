```python
df[(df.SepalWidthCm>4) | (df.SepalWidthCm<1)]
```

```python
df = df[(df.SepalWidthCm<4) & (df.SepalWidthCm>1)]
```

```python
df
```

```python
df.Species = df.Species.apply(lambda x: x.lstrip('Iris-'))
```

```python
df.Species.unique()
```

### Model Building (Naive Bayes)

```python
df
```

```python
x = df.drop(['Id','Species'], axis=1)
x
```

```python
y = df['Species'].replace(['setosa', 'versicolor', 'virginica'], [1,2,3])
y = y.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr, x_te, y_tr, y_te = train_test_split(x,y, test_size=0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
model1 = GaussianNB()
```

```python
model1.fit(x_tr,y_tr)
```

```python
y_pred = model1.predict(x_te)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_pred, y_te)
```

```python
confusion_matrix(y_pred, y_te)
```
