```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('adult1.xlsx')
df.head()
```

### Subset

```python
df['native-country'].unique()
```

```python
df['race'].unique()
```

```python
df['gender'].unique()
```

```python
sub1 = df[(df['gender']=='Male') & (df['race']=='White') & (df['native-country']=='China')]
```

```python
sub1
```

```python
sub2 = df[(df['gender']=='Male') & (df['race']=='Asian-Pac-Islander') & (df['native-country']=='China')]
```

```python
sub2
```

### Merge

```python
df_merged = pd.merge(sub1, sub2, how="outer")
```

```python
df_merged
```
